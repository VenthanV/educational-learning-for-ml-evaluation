```python
import pandas as pd
import numpy as np
import os
os.makedirs("scenario", exist_ok=True)
np.random.seed(42)

prob_safe_A = np.clip(np.random.normal(0.05, 0.05, 9900), 0.0, 1.0)
prob_mal_A = np.clip(np.random.normal(0.85, 0.1, 100), 0.0, 1.0)
df_ransomware = pd.DataFrame({
    'y_true': [0]*9900 + [1]*100,
    'y_prob': prob_safe_A.tolist() + prob_mal_A.tolist()
})
df_ransomware.to_csv("scenario/ransomware.csv", index=False)

prob_safe_B = np.clip(np.random.normal(0.3, 0.15, 9000), 0.0, 1.0)
prob_mal_B = np.clip(np.random.normal(0.6, 0.15, 1000), 0.0, 1.0)
df_firewall = pd.DataFrame({
    'y_true': [0]*9000 + [1]*1000,
    'y_prob': prob_safe_B.tolist() + prob_mal_B.tolist()
})
df_firewall.to_csv("scenario/firewall.csv", index=False)

prob_safe_C = np.clip(np.random.normal(0.2, 0.1, 5000), 0.0, 1.0)
prob_mal_C = np.clip(np.random.normal(0.8, 0.1, 5000), 0.0, 1.0)
df_phishing = pd.DataFrame({
    'y_true': [0]*5000 + [1]*5000,
    'y_prob': prob_safe_C.tolist() + prob_mal_C.tolist()
})
df_phishing.to_csv("scenario/phishing.csv", index=False)

prob_safe_D = np.clip(np.random.normal(0.2, 0.2, 10000), 0.0, 1.0)
prob_mal_D = np.clip(np.random.normal(0.9, 0.05, 50), 0.0, 1.0)
df_biometrie = pd.DataFrame({
    'y_true': [0]*10000 + [1]*50,
    'y_prob': prob_safe_D.tolist() + prob_mal_D.tolist()
})
df_biometrie.to_csv("scenario/biometrie.csv", index=False)

```
